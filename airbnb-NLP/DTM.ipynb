```python
ldamodel = gensim.models.ldamodel.LdaModel.load('pass_10_iter_3000.lda')
```

```python
ntopic = 6
aspect_kws = []
for i in range(ntopic) :
    kw = []
    kws_lst = ldamodel.show_topic(i, 100)
    for i in range(len(kws_lst)) :
        kws = kws_lst[i][0]
        kw.append(kws)
    aspect_kws.append(kw)
```

```python
words = []
for s in aspect_kws :
    words += s
freq = FreqDist(words)
aspect_vocab = []
for k,v in freq.items():
        aspect_vocab.append(k)
```

```python
review_room = []
for id, rew_lst in review_postag.items() :
    review_rev = []
    for rev in rew_lst :
        for sen in rev[1] :
            review_rev.append(sen) 
    review_room.append(review_rev)        
```

```python
vocab_dict = dict(zip(aspect_vocab, range(len(aspect_vocab))))
vocab_dict
```

```python
t = len(vocab_dict)
d = len(review_room)
print(t)
print(d)
```

```python
DTM = np.zeros((d,t))

for i in tqdm.tqdm(range(d)) :
    for w in review_room[i] :
        if w in vocab_dict:
            DTM[i][vocab_dict[w]] += 1
```

```python
import csv

with open("coef.csv") as f:
    reader = csv.reader(f)
    next(reader) 
    data = [r for r in reader]
```

```python
coef = []
for lst in data :
    coef.append(lst[1])  
```

```python
coeff = []
for p in coef :
    p = float(p)
    coeff.append(p)
```

```python
len(coeff)
```

```python
DCoef = np.zeros((t,1))

for i in range(t) :
    DCoef[i][0] = coeff[i]
```

```python
DCoef
```

```python
DTM_S = np.zeros((d,t))

for i in tqdm.tqdm(range(d)) :
    for j in range(t) :
        DTM_S[i][j] = DTM[i][j] * DCoef[j]
```
