### LRAR method 

```python
review_sent = []

for id,rev_lst in review.items() :
    for rev in rev_lst :
        review_sent.append(rev[1])
```

```python
only_sent = []

for id,rev_lst in review.items() :
    for rev in rev_lst :
        for s in rev[1] :
            only_sent.append(s)
```

```python
stopWords =list(stopwords.words('english'))
add_stopwords = ['airbnb','im',"i'm","we'll","i'll","i'd"'youre','hongdae', 'seoul', 'daegu', 'jeju' ,'busan', 'korea','gangneung','daejeon','yeosu',
                 'itaewon','myeongdong','john','gangnam','also','even','dont','namsan','incheon','jeonju','good','great','well','gwangju',"'s","...","'ve","``","''","'m",'--',"'ll","'d"]
# additional_stopwords = []
stopWords = set(stopWords + add_stopwords)
```

```python
words = []
for s in only_sent :
    words += s
freq = FreqDist(words)
vocab = []
for k,v in freq.items():
    if v > 99 :
        vocab.append(k)

vocab_dict = dict(zip(vocab, range(len(vocab))))
```

```python
len(vocab_dict)
```

```python
ntopic = 6
aspect_terms = []
for i in range(ntopic) :
    kw = []
    kws_lst = ldamodel.show_topic(i, 200)
    for i in range(len(kws_lst)) :
        kws = kws_lst[i][0]
        kw.append(kws)
    aspect_terms.append(kw)
```

```python
aspect_terms
```

```python
import numpy as np
```

```python
review_labels = []
k = ntopic # number of topic = 6
v = len(vocab) # 2589개

for r in tqdm.tqdm(review_sent) :  # 120383개의 리뷰에 대해
    labels = []
    for s in r:
        count = np.zeros(len(aspect_terms))
        i = 0
        for a in aspect_terms:
            for w in s:
                if w in vocab_dict :
                    if w in a:
                        count[i] += 1   # count : 문장 하나에서 aspect 별 단어의 freq
            i = i + 1

        if max(count) > 0:
            la = np.where(np.max(count) == count)[0].tolist() # 문장이 어떤 aspect 인지 labeling
            labels.append(la) 
        else:
            labels.append([])
    review_labels.append(labels)
```
